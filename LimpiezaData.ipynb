## **Introducción**

En el entorno industrial contemporáneo, la confiabilidad y la eficiencia de los activos operativos, ya sean máquinas rotativas, sistemas hidráulicos, sistemas eléctricos, o cualquier otro sistema que pueda ser monitoreado mediante mediciones, son elementos críticos para asegurar la continuidad operativa y la optimización del mantenimiento. La capacidad de predecir y prevenir fallas basada en datos reales, así como de detectar anomalías en los parámetros operativos, permite anticipar eventos adversos, minimizar tiempos de inactividad, reducir costos, prolongar la vida útil de los activos, y mantener estándares óptimos de calidad y seguridad.

Este estudio propone un sistema robusto y adaptativo de prognosis industrial enfocado en la detección temprana de anomalías y la predicción de fallas para cualquier activo cuantificable. El enfoque es híbrido, combinando técnicas estadísticas tradicionales con métodos de machine learning. La metodología incluye desde el preprocesamiento de grandes volúmenes de datos de medición, la identificación de variables críticas, el establecimiento de líneas base adaptativas, hasta la generación de alertas predictivas con objetivos operativos claros.

El documento detalla de manera preliminar con un ejemplo, el sistema de prognosis, partiendo del análisis de un conjunto de datos industriales de amplio espectro, pasando por las fases de preprocesamiento y limpieza de datos, hasta un análisis exploratorio inicial. Se introduce la importancia de las matrices de varianza-covarianza y correlación para entender relaciones entre variables. Se compara la estrategia seleccionada de variables con técnicas reconocidas como el Análisis de Componentes Principales (ACP) y el Análisis Factorial, evaluando su idoneidad para series temporales industriales y para activos cuya falla se manifiesta gradualmente o presenta patrones estocásticos.


## **Objetivo:**

Esta sección del trabajo tiene como objetivo general explorar la estructura subyacente y las interrelaciones de las variables de calidad de energía en la Subestación Canaima, utilizando técnicas de análisis multivariado para la reducción de dimensionalidad y la interpretación de patrones complejos, en preparación para fases posteriores del sistema de prognosis.

### **Contexto del Conjunto de Datos y Objetivo del Análisis:**

El sistema de prognosis se fundamenta en un conjunto de datos cuantitativos recopilado de una subestación de energía industrial: la Subestación Canaima de 1000kVA. Variables Objetivos:
Basados en los criterios tecnicos de expertos, las variables eléctricas clave que pueden ser las más representativas:

Tensiones de línea y fase (Tensión L1, L2, L3, y sus mínimas y máximas):

Razón: La estabilidad y calidad del suministro eléctrico depende directamente de la tensión. Cualquier desviación significativa en las tensiones, especialmente si se sale de los rangos permitidos, puede indicar problemas en el sistema, como sobrecargas, desequilibrio de fases, o incluso falla en transformadores.
Importancia: Las tensiones fuera de los rangos normales pueden generar sobrecalentamiento o desbalance que impacta en la eficiencia y puede ser indicativo de fallas inminentes.

Corrientes de fase (Corriente L1, L2, L3):

Razón: Las corrientes reflejan directamente el consumo de energía y las cargas conectadas al sistema. Un incremento o fluctuación en las corrientes puede estar relacionado con sobrecargas, cortocircuitos o fallas en los equipos de la subestación.
Importancia: Es esencial monitorear estas corrientes para detectar cualquier sobrecarga que podría derivar en un fallo.

Potencias (Aparente, Activa, Reactiva):

Razón: La potencia activa (kW) es la que realmente realiza trabajo útil, mientras que la potencia reactiva (kVAR) es necesaria para mantener los campos eléctricos y magnéticos en los equipos. Un desbalance entre estos dos tipos de potencias puede indicar problemas en la eficiencia energética o fallas en los equipos que generen exceso de consumo de energía reactiva.
Importancia: El exceso de potencia reactiva puede sobrecargar los transformadores y sistemas de distribución, mientras que la potencia activa refleja el trabajo real. El monitoreo de la potencia total (aparente) permite observar posibles sobrecargas.

Factor de potencia (FP):

Razón: Un factor de potencia bajo significa que se está consumiendo una gran cantidad de potencia reactiva en comparación con la potencia activa, lo cual es ineficiente y puede causar sobrecargas en el sistema.
Importancia: Es un indicador importante de eficiencia. Valores bajos deben corregirse y pueden estar asociados a fallas en los equipos de corrección del factor de potencia, como los condensadores.

Distorsión armónica (THD) en tensión y corriente:

Razón: La distorsión armónica indica la presencia de armónicos (frecuencias no deseadas) en la señal de tensión o corriente, que pueden provenir de equipos electrónicos no lineales o cargas como motores de velocidad variable o rectificadores. Estos armónicos causan pérdidas adicionales y posibles fallas en los transformadores y otros equipos.
Importancia: Niveles elevados de distorsión armónica pueden generar un sobrecalentamiento en los transformadores, aumentar las pérdidas y reducir la vida útil del equipo.

Temperatura de los transformadores y equipos:

Razón: Los sistemas eléctricos se ven afectados por el calor, especialmente los transformadores, que pueden sobrecalentarse bajo carga. Monitorear la temperatura ayuda a prevenir fallas térmicas.
Importancia: Las temperaturas fuera de rango pueden indicar problemas como sobrecarga, aislamiento defectuoso o mal funcionamiento de los sistemas de enfriamiento, para nuestro caso no hace parte del estudio pero se menciona con fines de indicar la importancia de esta variable medida en los sitemas eslectricos y la necesidad de ser incluida para este tipo de analisis.  

Frecuencia del sistema:

Razón: La frecuencia debe mantenerse constante en el sistema de distribución (normalmente alrededor de 50 o 60 Hz, dependiendo de la región). Cambios significativos en la frecuencia pueden ser una indicación de desbalance entre generación y consumo, o problemas serios en el sistema.
Importancia: Una frecuencia fuera del rango normal puede causar problemas en equipos sensibles y reflejar fallas en el sistema de control de la subestación.

Variables secundarias que podrían no ser prioritarias:

Armónicos de órdenes superiores (> 15): Si bien los armónicos de orden bajo pueden tener mayor impacto, los de órdenes más altos (20, 30, 40, etc.) pueden no ser tan críticos para la predicción de fallas en subestaciones.

Variables de eventos específicos: Dependiendo del sistema, algunos eventos puntuales o picos pueden ser menos importantes si ocurren con poca frecuencia o no afectan el consumo energético general.


Las variables que reflejan el comportamiento general del sistema eléctrico, como tensiones, corrientes, potencias, distorsión armónica y factor de potencia, segun los criterios tecnicos mas comunes y con el fin de reducir la dimensionalidad de nuestro estudio, de determina cuales son las más importantes para la predicción de fallas. Variables relacionadas con condiciones específicas, como armónicos de orden alto o eventos puntuales, podrían ser consideradas secundarias y podrían eliminarse o reducirse en el análisis.

Según Daniel Peña (cap. 4.3.1 4.3.2, 4.3.3 y 5.5)

Si las variables están en distintas escalas → estandarizar (media 0, varianza 1) y trabajar con la matriz de correlaciones.

Si las variables están en la misma escala y la varianza es informativa → usar directamente la matriz de covarianzas.

Estrategia recomendada 

Estandarizar todas las variables (restar la media y dividir por la desviación típica).

Esto elimina el sesgo de escala.lo realizaremos con StandardScaler() en el preprocesamiento.

Usar la matriz de correlaciones (Pearson) como insumo para el ACP y el AF.

Así evitaremos que las tensiones (que tienen valores mucho más grandes) dominen sobre armónicos o factores de potencia.

Reducir dimensionalidad previa:

Eliminar una de cada par de variables con correlaciones altísimas (>0.98 o 0.99) para evitar redundancia. Esto simplifica la interpretación posterior de componentes y factores.

La mejor estrategia es normalizar con StandardScaler y trabajar con la matriz de correlaciones para el ACP y el AF.



```python
# CELDA: Preprocesamiento local (DataFrame-only)
# - Detecta y elimina variables constantes / muy poco informativas
# - Calcula scores por variable (varianza, diversidad, estabilidad, tendencia, correlación)
# - Elimina variables redundantes por alta correlación (umbral configurable)
# - Estandariza (Z-score) las variables seleccionadas
# - Guarda reportes en ./eda_outputs

import os
import numpy as np
import pandas as pd
import unicodedata
import logging
from typing import Dict, Any, List, Tuple
from sklearn.preprocessing import StandardScaler

OUTPUT_DIR = "/home/elicoubuntu/Desktop/EstadisticaMultivariable/eda_outputs"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# --- Configuración de logging simple ---
logger = logging.getLogger("preprocessor")
if not logger.handlers:
    h = logging.StreamHandler()
    h.setFormatter(logging.Formatter("%(asctime)s - %(levelname)s - %(message)s"))
    logger.addHandler(h)
logger.setLevel(logging.INFO)


class DataFramePreprocessor:
    def __init__(
        self,
        file_path: str = None,
        timestamp_col: str = "timestamp",
        corr_threshold: float = 0.98,
        min_unique: int = 3,
        unique_ratio_threshold: float = 0.001,
        variance_threshold: float = 1e-6,
    ):
        self.file_path = file_path
        self.timestamp_col = timestamp_col
        self.corr_threshold = corr_threshold
        self.min_unique = min_unique
        self.unique_ratio_threshold = unique_ratio_threshold
        self.variance_threshold = variance_threshold

        self.raw = None
        self.cleaned = None
        self.variable_mapping: Dict[str, str] = {}
        self.scores: Dict[str, Dict[str, float]] = {}
        self.removed_variables: List[Dict[str, Any]] = []
        self.selected_variables: List[str] = []
        self.scaler = StandardScaler()

    # -------------------------
    # Utilidades
    # -------------------------
    def _normalize_column_name(self, col: str) -> str:
        s = col.lower()
        s = unicodedata.normalize("NFKD", s)
        s = "".join([c for c in s if not unicodedata.combining(c)])
        s = "".join(c if c.isalnum() else "_" for c in s)
        s = "_".join(filter(None, s.split("_")))
        # keep mapping
        self.variable_mapping[s] = col
        return s

    def load_data(self) -> pd.DataFrame:
        """Carga CSV o Excel si se pasa file_path; si self.raw ya existe lo deja."""
        if self.raw is not None:
            return self.raw
        if not self.file_path:
            raise ValueError("Proporcione file_path al inicializar el preprocesador")
        ext = self.file_path.split(".")[-1].lower()
        logger.info(f"Cargando archivo {self.file_path} ...")
        if ext in ("xlsx", "xls"):
            df = pd.read_excel(self.file_path)
        elif ext == "csv":
            df = pd.read_csv(self.file_path)
        else:
            raise ValueError("Formato no soportado. Use .xlsx/.xls o .csv")
        self.raw = df
        logger.info(f"Datos cargados: {df.shape[0]} filas x {df.shape[1]} columnas")
        return df

    # -------------------------
    # Limpieza básica
    # -------------------------
    def clean_data(self, df: pd.DataFrame = None) -> Tuple[pd.DataFrame, Dict[str, Any]]:
        """Normaliza nombres, elimina variables constantes, rellena NaNs y reporta."""
        if df is None:
            df = self.load_data()
        df = df.copy()
        # normalizar nombres
        df.columns = [self._normalize_column_name(c) for c in df.columns]

        report = {
            "initial_shape": (df.shape[0], df.shape[1]),
            "constant_variables": [],
            "missing_values": {},
            "data_types": {},
        }

        # detectar constantes (≤1 valor único)
        const_vars = []
        for c in df.columns:
            nuniq = df[c].nunique(dropna=True)
            if nuniq <= 1:
                const_vars.append({"variable": c, "value": df[c].iloc[0] if not df[c].empty else None, "n_unique": nuniq})
        report["constant_variables"] = const_vars

        # eliminar constantes
        if const_vars:
            const_cols = [v["variable"] for v in const_vars]
            logger.warning(f"Eliminando variables constantes: {const_cols}")
            df.drop(columns=const_cols, inplace=True)

        # registrar tipos y nulos; rellenar nulos (num -> median, cat -> 'Desconocido')
        for c in df.columns:
            report["data_types"][c] = str(df[c].dtype)
            nnull = int(df[c].isnull().sum())
            report["missing_values"][c] = nnull
            if nnull > 0:
                if pd.api.types.is_numeric_dtype(df[c]):
                    med = df[c].median()
                    df[c].fillna(med, inplace=True)
                else:
                    df[c].fillna("Desconocido", inplace=True)

        # intentar parsear timestamp si existe
        if self.timestamp_col in df.columns:
            try:
                df[self.timestamp_col] = pd.to_datetime(df[self.timestamp_col])
            except Exception:
                logger.warning(f"No se pudo convertir columna {self.timestamp_col} a datetime; se conserva")
        else:
            logger.info("No se encontró columna timestamp (no es obligatorio).")

        self.cleaned = df
        logger.info(f"Limpieza completada. Forma resultante: {df.shape[0]} x {df.shape[1]}")
        return df, report

    # -------------------------
    # Scoring y detección de variables inútiles
    # -------------------------
    def _calculate_variable_score(self, series: pd.Series, data_numeric: pd.DataFrame) -> Dict[str, float]:
        """Calcula métricas y devuelve dict con scores parciales y final."""
        s = series.dropna()
        n = len(s)
        res = {
            "n_unique": int(series.dropna().nunique()),
            "unique_ratio": 0.0,
            "variance": float(series.var(ddof=1)) if n > 1 else 0.0,
            "variance_score": 0.0,
            "stability_score": 0.0,
            "trend_score": 0.0,
            "correlation_score": 0.0,
            "final_score": 0.0,
        }

        if n == 0:
            return res

        n_unique = res["n_unique"]
        res["unique_ratio"] = n_unique / n if n > 0 else 0.0

        # quick filters
        if n_unique <= self.min_unique:
            return res
        if res["unique_ratio"] < self.unique_ratio_threshold:
            return res

        var = res["variance"]
        if abs(var) < self.variance_threshold:
            return res

        # variance_score (log-scaled * diversity)
        res["variance_score"] = float(np.log1p(abs(var)) * res["unique_ratio"])

        # stability_score: windowed rolling std vs total std (if timestamp indexed better)
        try:
            ser = series.copy()
            if self.timestamp_col in self.cleaned.columns and pd.api.types.is_datetime64_any_dtype(self.cleaned[self.timestamp_col]):
                # reindex by time for rolling
                tmp = pd.concat([self.cleaned[self.timestamp_col], ser], axis=1).set_index(self.timestamp_col)
                rolling_std = tmp[ser.name].rolling("1H").std().mean()  # 1-hour window as heuristic
                total_std = ser.std()
            else:
                rolling_std = ser.rolling(window=24, min_periods=1).std().mean()
                total_std = ser.std()
            if total_std > 0 and not np.isnan(rolling_std):
                stability = 1.0 - (rolling_std / total_std)
                res["stability_score"] = float(max(0.0, min(1.0, stability)))
        except Exception:
            res["stability_score"] = 0.0

        # trend_score: absolute slope of simple linear fit (normalized by std to be scale-invariant)
        try:
            y = series.dropna().values
            x = np.arange(len(y))
            if len(y) > 1:
                slope = np.polyfit(x, y, 1)[0]
                # normalize slope by range or std:
                denom = (series.std() if series.std() > 0 else 1.0)
                res["trend_score"] = float(abs(slope) / denom)
        except Exception:
            res["trend_score"] = 0.0

        # correlation_score: mean absolute correlation with other numeric variables
        try:
            if data_numeric is not None and data_numeric.shape[1] > 1:
                corrs = data_numeric.corrwith(series).abs().drop(labels=[series.name], errors="ignore")
                if not corrs.empty:
                    res["correlation_score"] = float(corrs.mean())
        except Exception:
            res["correlation_score"] = 0.0

        # final weighted score
        weights = {"variance_score": 0.35, "stability_score": 0.25, "trend_score": 0.2, "correlation_score": 0.2}
        final = sum(res[k] * w for k, w in weights.items())
        res["final_score"] = float(final)
        return res

    def analyze_variables(self) -> Dict[str, Any]:
        """Analiza todas las variables numéricas y decide descartes iniciales."""
        if self.cleaned is None:
            raise ValueError("Ejecuta clean_data() antes de analyze_variables()")

        df = self.cleaned.copy()
        numeric = df.select_dtypes(include=[np.number])
        logger.info(f"Analizando {numeric.shape[1]} variables numéricas para scoring...")

        scores = {}
        removed = []
        for col in numeric.columns:
            res = self._calculate_variable_score(numeric[col], numeric)
            scores[col] = res
            # if final_score == 0 => variable a remover por falta de información
            if res["final_score"] == 0.0:
                removed.append({"variable": col, "reason": "score_zero", **res})

        # registrar
        self.scores = scores
        self.removed_variables = removed

        # crear DataFrame de scores y guardarlo
        scores_df = pd.DataFrame(scores).T
        scores_df.to_csv(os.path.join(OUTPUT_DIR, "variable_scores.csv"))
        pd.DataFrame(removed).to_csv(os.path.join(OUTPUT_DIR, "removed_by_score.csv"), index=False)

        logger.info(f"Variables marcadas como eliminadas por score: {len(removed)}")
        return {"scores_df_path": os.path.join(OUTPUT_DIR, "variable_scores.csv"),
                "removed_by_score_path": os.path.join(OUTPUT_DIR, "removed_by_score.csv"),
                "n_removed_initial": len(removed)}

    # -------------------------
    # Reducción por correlación
    # -------------------------
    def remove_highly_correlated(self, keep_strategy: str = "highest_score") -> Dict[str, Any]:
        """
        Elimina una variable de cada par con |corr| >= corr_threshold.
        keep_strategy='highest_score' -> mantiene la variable con mayor final_score
        """
        if self.scores is None or not self.scores:
            raise ValueError("Ejecuta analyze_variables() antes de remove_highly_correlated()")

        # start from numeric vars that survived initial removal
        numeric = self.cleaned.select_dtypes(include=[np.number]).columns.tolist()
        removed_initial = {r["variable"] for r in self.removed_variables}
        candidates = [c for c in numeric if c not in removed_initial]

        if len(candidates) <= 1:
            logger.info("Pocas variables candidatas para correlational pruning.")
            self.selected_variables = candidates
            return {"selected": self.selected_variables, "n_removed_corr": 0}

        corr = self.cleaned[candidates].corr().abs()
        # turn into pairs sorted by corr descending
        pairs = (
            corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
            .stack()
            .sort_values(ascending=False)
        )

        to_remove = set()
        for (v1, v2), corrval in pairs.items():
            if corrval < self.corr_threshold:
                break  # subsequent pairs will be smaller
            if v1 in to_remove or v2 in to_remove:
                continue
            # choose which to remove
            score_v1 = self.scores.get(v1, {}).get("final_score", 0.0)
            score_v2 = self.scores.get(v2, {}).get("final_score", 0.0)
            if keep_strategy == "highest_score":
                if score_v1 >= score_v2:
                    rm = v2
                else:
                    rm = v1
            else:
                # default: remove v2
                rm = v2
            to_remove.add(rm)

        # register removals
        removed_corr = [{"variable": v, "reason": f"corr>={self.corr_threshold}"} for v in sorted(list(to_remove))]
        # drop them from cleaned copy
        final_candidates = [c for c in candidates if c not in to_remove]

        # update attributes
        self.removed_variables.extend(removed_corr)
        self.selected_variables = final_candidates

        # save reports
        pd.DataFrame(removed_corr).to_csv(os.path.join(OUTPUT_DIR, "removed_by_correlation.csv"), index=False)
        pd.DataFrame({"selected_variables": self.selected_variables}).to_csv(os.path.join(OUTPUT_DIR, "selected_variables.csv"), index=False)

        logger.info(f"Eliminadas {len(removed_corr)} variables por alta correlación. Quedan {len(final_candidates)} variables.")
        return {"removed_by_correlation_path": os.path.join(OUTPUT_DIR, "removed_by_correlation.csv"),
                "selected_variables_path": os.path.join(OUTPUT_DIR, "selected_variables.csv"),
                "n_removed_corr": len(removed_corr),
                "n_selected": len(final_candidates)}

    # -------------------------
    # Normalización final (z-score) sobre variables seleccionadas
    # -------------------------
    def normalize_selected(self) -> pd.DataFrame:
        if not self.selected_variables:
            numeric = self.cleaned.select_dtypes(include=[np.number]).columns.tolist()
            removed_initial = {r["variable"] for r in self.removed_variables}
            self.selected_variables = [c for c in numeric if c not in removed_initial]

        df = self.cleaned.copy()
        cols = self.selected_variables
        if not cols:
            raise ValueError("No hay variables seleccionadas para normalizar")

        logger.info(f"Estandarizando {len(cols)} variables seleccionadas...")
        arr = self.scaler.fit_transform(df[cols].values)

        # ⚡ Nuevo: crear DataFrame solo con seleccionadas (+ timestamp si existe)
        df_norm = pd.DataFrame(arr, columns=cols, index=df.index)
        if self.timestamp_col in df.columns:
            df_norm.insert(0, self.timestamp_col, df[self.timestamp_col].values)

        # guardar solo seleccionadas
        path = os.path.join(OUTPUT_DIR, "normalized_selected_data.csv")
        df_norm.to_csv(path, index=False)
        logger.info(f"Normalized data saved to {path}")
        return df_norm



# -------------------------
# EJEMPLO DE USO 
# -------------------------
if __name__ == "__main__":
    FILE = "/home/elicoubuntu/Desktop/EstadisticaMultivariable/filtered_consolidated_data_cleaned.xlsx"
    pre = DataFramePreprocessor(file_path=FILE, corr_threshold=0.98)
    raw = pre.load_data()
    cleaned, report = pre.clean_data(raw)
    logger.info("Reporte limpieza (ejemplo):")
    # Imprime un resumen conciso
    logger.info({k: (report[k] if k in report else None) for k in ["initial_shape", "constant_variables"]})
    pre.analyze_variables()
    pre.remove_highly_correlated()
    norm_df = pre.normalize_selected()

    logger.info("Resumen final:")
    logger.info(f"Filas: {norm_df.shape[0]}, Columnas (normalizadas): {len(pre.selected_variables)}")
    logger.info(f"Reportes guardados en: {OUTPUT_DIR}")
```

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA, FactorAnalysis

# === 1. Cargar datos normalizados ===
data_path = "/home/elicoubuntu/Desktop/EstadisticaMultivariable/eda_outputs/normalized_selected_data.csv"
df = pd.read_csv(data_path)

print(f"Datos cargados: {df.shape[0]} filas x {df.shape[1]} columnas")

# === 2. Filtrar solo columnas numéricas ===
numeric_df = df.select_dtypes(include=[np.number])
print(f"Columnas numéricas seleccionadas: {numeric_df.shape[1]}")

# === 3. PCA ===
pca = PCA()
pca.fit(numeric_df)

eigenvalues = pca.explained_variance_
n_components = sum(eigenvalues > 1)  # criterio de Kaiser
pcs = pca.transform(numeric_df)[:, :n_components]

pcs_df = pd.DataFrame(
    pcs,
    columns=[f"PC{i+1}" for i in range(n_components)],
    index=numeric_df.index
)

print("\n--- Puntajes (PCs) - Primeras 20 filas ---")
print(pcs_df.head(20).to_markdown(floatfmt=".4f"))

# Guardar PCs completos
pcs_df.to_csv("/home/elicoubuntu/Desktop/EstadisticaMultivariable/eda_outputs/puntajes_pca.csv")

# === 4. AF ===
fa = FactorAnalysis(n_components=n_components, random_state=42)
factors = fa.fit_transform(numeric_df)

factors_df = pd.DataFrame(
    factors,
    columns=[f"Factor{i+1}" for i in range(n_components)],
    index=numeric_df.index
)

print("\n--- Puntajes (Factores AF) - Primeras 20 filas ---")
print(factors_df.head(20).to_markdown(floatfmt=".4f"))

# Guardar Factores completos
factors_df.to_csv("/home/elicoubuntu/Desktop/EstadisticaMultivariable/eda_outputs/puntajes_factores.csv")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA, FactorAnalysis

# === 1. Cargar datos normalizados ===
data_path = "/home/elicoubuntu/Desktop/EstadisticaMultivariable/eda_outputs/normalized_selected_data.csv"
df = pd.read_csv(data_path)

numeric_df = df.select_dtypes(include=[np.number])

# === 2. PCA ===
pca = PCA()
pca.fit(numeric_df)
pcs = pca.transform(numeric_df)

# === 3. Factor Analysis (con n_factors=6 por Kaiser) ===
fa = FactorAnalysis(n_components=6, random_state=42)
factors = fa.fit_transform(numeric_df)

# === 4. Función para graficar biplots ===
def biplot(scores, loadings, labels, dim1, dim2, title, color):
    plt.figure(figsize=(8,6))
    # puntos (observaciones)
    plt.scatter(scores[:, dim1], scores[:, dim2], alpha=0.3)
    # vectores (variables)
    for i, var in enumerate(labels):
        plt.arrow(0, 0, loadings[i, dim1]*5, loadings[i, dim2]*5, 
                  color=color, alpha=0.7, head_width=0.05)
        plt.text(loadings[i, dim1]*5.2, loadings[i, dim2]*5.2, var, 
                 color=color, ha='center', va='center')
    plt.xlabel(f"Dim {dim1+1}")
    plt.ylabel(f"Dim {dim2+1}")
    plt.title(title)
    plt.grid(True)
    plt.show()

# === 5. Biplots PCA (primeros 6 PCs) ===
loadings_pca = pca.components_.T
labels = numeric_df.columns

biplot(pcs, loadings_pca, labels, 0, 1, "Biplot PCA (PC1 vs PC2)", "red")
biplot(pcs, loadings_pca, labels, 2, 3, "Biplot PCA (PC3 vs PC4)", "red")
biplot(pcs, loadings_pca, labels, 4, 5, "Biplot PCA (PC5 vs PC6)", "red")

# === 6. Biplots FA (primeros 6 factores) ===
loadings_fa = fa.components_.T

biplot(factors, loadings_fa, labels, 0, 1, "Biplot FA (Factor1 vs Factor2)", "green")
biplot(factors, loadings_fa, labels, 2, 3, "Biplot FA (Factor3 vs Factor4)", "green")
biplot(factors, loadings_fa, labels, 4, 5, "Biplot FA (Factor5 vs Factor6)", "green")
```

Comparación ACP vs AF

ACP (PCA):

Busca maximizar la varianza explicada con el menor número de componentes. Cada componente es combinación lineal de todas las variables originales. No supone un modelo subyacente, solo es una transformación geométrica.

AF (Factor Analysis):

Busca identificar factores latentes (no observados) que explican las correlaciones entre variables.

Se centra en la covarianza común y descarta la varianza específica y el error.

Es un modelo probabilístico con interpretación más “teórica”.

Número de dimensiones: 

En ACP, los primeros 6 componentes capturaron la mayor parte de la varianza (según regla de Kaiser y varianza acumulada).

En AF, también obtuvimos 6 factores significativos.

Ambos métodos convergen en que 6 dimensiones son suficientes para representar gran parte de la información.

Puntajes

Los puntajes reflejan la influencia de factores latentes sobre cada observación.

Muestran tendencias similares al ACP, pero con mayor interpretabilidad respecto a grupos de variables.

Biplots

PCA:Se observó una clara dispersión y agrupamiento de observaciones en los primeros componentes.
Las variables proyectadas en los ejes principales muestran cuáles dominan cada dimensión (ej: variables de corriente y tensión muy asociadas a PC1 y PC2).

AF:Los factores agruparon mejor las variables correlacionadas (ej: armónicos, distorsiones, tensiones).Esto facilita una interpretación conceptual, donde cada factor representa un conjunto de variables relacionadas.

Interpretación

PCA es más útil para reducción de dimensionalidad y visualización, pero sus componentes no siempre tienen una interpretación clara, el ACP no intenta descubrir causas ocultas o fenómenos detrás de los datos, simplemente reorganiza la información para que sea más fácil de visualizar y resumir.

AF permite dar sentido a los factores (ej: un factor podría representar “calidad de tensión”, otro “efectos de armónicos”, otro “corriente”), lo cual es más útil si el interés es explicar las relaciones subyacentes entre variables, el AF sí asume que detrás de las variables medidas hay causas ocultas que no vemos directamente.
